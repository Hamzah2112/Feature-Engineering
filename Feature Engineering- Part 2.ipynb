## Encoding Techniques

Encoding techniques are performed on categorical features and there are 2 types of categorical features:-

#### **1-Nominal Encoding** :- 
In this type of encoding we do not worry about ranking or the sequence of the categories.They are of following types:-


(a)- **One Hot Encoding**:- In this type of coding technique we do not require to rank the categories. So we can replace the categories by dummy feature

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('titanic.csv',usecols=['Sex'])
df.head()
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

(b)- **One hot encoding with many categories in a feature**:-In this type of coding technique we use 10 most frequent labels and convert them into dummy variables using one hot encoding.

```python
df=pd.read_csv('mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
df.head()
```

```python
# We can see for different features there are the count of diffrent labels are
for i in df.columns:
    print(len(df[i].unique()))
```

```python
#We are considering X1 feature
df['X1'].value_counts().sort_values(ascending=False)
```

```python
# We took top 10 values for categories and put it in the list
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
lst_10
```

```python
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```

```python
lst_10.append('X1')
```

```python
df[lst_10].head()
```

```python
df.head()
```

#### Advantages
* Easy To Use
* works well when there are less label in the feature

#### Disadvantages
* Increases the size of feature

(c)- **Count Or Frequency Encoding**:-In this type of coding technique we replace the categorical features with the count/frequence of each label in the paticular feature.  

```python
df=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
df.head()
```

```python
## Selecting categorical features from the dataset
columns=[1,3,5,6,7,8,9,13]
data=df[columns]
data.head()
```

```python
## Providing names to the feature(since columns names are not present)
data.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
data.head()
```

```python
## Finding unique labels in the features
for feature in data.columns[:]:
    print(feature,'have',len(data[feature].unique()),'labels')
```

```python
## Applying count encoding to the country feature
country_count=data['Country'].value_counts().to_dict()
country_count
```

```python
data['Country']=data['Country'].map(country_count)
data.head()
```

##### Advantages

1. Easy To Use
2. Not increasing feature space

##### Disadvantages

1. It will provide same weight if the frequencies are same

### **2-Ordinal Encoding** :-
In this type of encoding we need to **rank or order** the category.They are of following types:-

(a)- **Ordinal Number Encoding**:- In this type of coding technique we need  to rank the categories by assigning a unique value to all the feature.

```python
import datetime
```

```python
days=datetime.datetime.today()
days
```

```python
## List comprehension
list_of_days=[days-datetime.timedelta(x) for x in range(0,15)]
```

```python
list_of_days
```

```python
## Converting list into dataframe
df=pd.DataFrame(list_of_days)
df.columns=['Days']
df.head()
```

```python
## adding an additionl column for weekday name
df['Weekday']=df['Days'].dt.day_name()
df.head()
```

```python
## Creating a dictonary to rank the number of days
dic={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
df['weekday_ordinal']=df['Weekday'].map(dic)
```

```python
## new column is cretaed in which ordinal encoding is done
df.head(10)
```

(b)- **Target guided Ordinal Encoding**:-In this type of coding technique we need  to rank the categories by ordering the labels according to the target and replace the labels by the joint probability of being 1 or 0

```python
dataset=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
dataset.head()
```

```python
dataset['Cabin'].fillna('Missing',inplace=True)
dataset.head()
```

```python
## We have taken first initial of categorical feature
dataset['Cabin']=dataset['Cabin'].astype(str).str[0]
dataset.head()
```

```python
dataset['Cabin'].unique()
```

```python
## ordering the labels according to the target
dataset.groupby(dataset['Cabin'])['Survived'].mean()
```

```python
ordinal_label=dataset.groupby(dataset['Cabin'])['Survived'].mean().sort_values().index
ordinal_label
```

```python
## Making a dictonary to assign values based on the target
ordinal_label_2={k:i for i,k in enumerate(ordinal_label,0)}
ordinal_label_2
```

```python
## replace the labels by the joint probability
dataset['Ordinal Label']=dataset['Cabin'].map(ordinal_label_2)
```

```python
dataset.head(10)
```

(c)- **Mean Encoding**:-In this type of coding technique we need  replace the lables in the categories by mean values of the labels.

```python
dataset.head()
```

```python
mean_ordinal=dataset.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
dataset['Mean Ordinal']=dataset['Cabin'].map(mean_ordinal)
```

```python
dataset.head()
```

(d)- **Probablity Ratio Encoding**:- In this type of coding technique we need replace the lables in the categories by probablity of occurance of the labels.
We follow the following steps:-
* Probability of Survived based on Cabin ------ Categorical Feature
* Probability of Not Survived            ------ **1 - Prob(Survived)**
* **Prob(Survived) / Prob(Not Survived)**
* Dictonary to map cabin with probability
* Replace with the categorical feature

```python
df=pd.read_csv('titanic.csv',usecols=['Survived','Cabin'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df['Cabin'].unique()
```

```python
prob_df=df.groupby(['Cabin'])["Survived"].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived']
prob_df
```

```python
prob_df['Probablity Ratio']=prob_df['Survived']/prob_df['Died']
prob_df
```

```python
prob_encoding=prob_df['Probablity Ratio'].to_dict()
```

```python
df['cabin_encoding']=df['Cabin'].map(prob_encoding)
```
